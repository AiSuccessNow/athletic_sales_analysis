#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
# Inner_joint was used to combine df2020 and df2021 to get combined_df 

combined_df = pd.concat([df2020, df2021], ignore_index=True)
combined_df.reset_index(drop=True, inplace=True)
combined_df
```

```python
# check if any value is null.
missing_count = combined_df.isnull().sum()
print(missing_count)
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])
combined_df.rename(columns={"invoice_date": "datetime"}, inplace=True)
combined_df
```

```python
#....Confirm invoice Date changed to datetime successfully
combined_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number of products sold (TOTAL PRODUCTS SOLD ) for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold_df = combined_df.groupby(['region','state','city'])['units_sold'].agg(Total_Products_Sold =('sum'))

# Show the top 5 results.
total_products_sold_df.sort_values(by = (["Total_Products_Sold"]),ascending=False).head()
```

#### Using `pivot_table`

```python
### Pivot Table
pivot_city_short_form = pd.pivot_table(combined_df, index=["region","state","city"], values="units_sold", aggfunc= "sum")
pivot_city_short_form.rename(columns={"units_sold":"Units_Sold"}, inplace=True)
pivot_city_short_form.sort_values(by = (["Units_Sold"]),ascending=False).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales (TOTAL SALES) for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_df = combined_df.groupby(['region','state','city'])['total_sales'].agg(Total_Sales=('sum'))
total_sales_df
# Show the top 5 results.
total_sales_df.sort_values("Total_Sales",ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"

pivot_city_short_form = pd.pivot_table(combined_df, index=["region","state","city"], values="total_sales", aggfunc= "sum")
pivot_city_short_form.rename(columns={"total_sales":"Total_Sales"}, inplace=True)
# Show the top 5 results.
pivot_city_short_form.sort_values(by=(["Total_Sales"]),ascending=False).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold (TOTAL SALES)for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_products_sold_df = combined_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].agg(Total_Sales =('sum'))

# Show the top 5 results.
total_products_sold_df.sort_values("Total_Sales",ascending=False).head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
unique_values = combined_df['product'].unique()
unique_values
```

```python
# #..... RENAMING 'datetime' back to >> 'invoice_date' 
# womens_df.rename(columns={"datetime":"invoice_date"}, inplace=True)
# womens_df.head(1)
womens_df.rename(columns={"datetime":"invoice_date"}, inplace=True)
womens_df.head(5)
```

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.
pivot_invoice_date_short_form = pd.pivot_table(womens_df, index=["invoice_date"], values="total_sales", aggfunc= "sum")
pivot_invoice_date_short_form.sort_values(by = (["total_sales"]),ascending=False)
pivot_invoice_date_short_form.rename(columns={"total_sales":"Total_Sales"}, inplace=True)
pivot_invoice_date_short_form.head(5)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
#weekly_resampled_data = df.open.resample('W').mean()
daily_pivot_invoice_date_short_form = pivot_invoice_date_short_form.resample('D').sum() #mean()
#daily_pivot_invoice_date_short_form
daily_pivot_invoice_date_short_form.sort_values("Total_Sales", ascending=False).head()
# Sort the resampled pivot table in descending order on "Total Sales".
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
weekly_pivot_invoice_date_short_form = pivot_invoice_date_short_form.resample('W').sum() #mean()
#daily_pivot_invoice_date_short_form
weekly_pivot_invoice_date_short_form.sort_values("Total_Sales", ascending=False).head()
# Sort the resampled pivot table in descending order on "Total Sales".
```
